```python
import pandas as pd
import math
```

```python
df = pd.read_csv("teste_21.csv", delimiter = ";")
```

```python
df.head()
```

```python
def minimo(a):
    if (a['30h'] > 0 and a['06h'] == 0) or (a['30h'] == 0 and a['06h'] > 0):
        if a['30h'] + a['06h'] < 4:
            return 'Minimo'
        else:
            return "Espaco"
    else:
        if a['30h'] + a['06h'] < 5:
            return "Minimo"
        else:
            return "Espaco"
```

```python
df['minimo'] = df.apply(minimo, axis=1)
```

```python
df.head()
```

```python
df['delta_30'] = df['30h_d'] - df['30h']
df['delta_06'] = df['06h_d'] - df['06h']
df['delta_final'] = (df['30h_d'] + df['06h_d']) - (df['30h'] + df['06h']) 
```

```python
df.head()
```

```python
def status(a):
    if a > 0:
        return "Acrescimo"
    elif a < 0:
        return "Reducao"
    else:
        return "Quadro Ideal"
```

```python
df['status_30'] = df['delta_30'].apply(status)
df['status_06'] = df['delta_06'].apply(status)
df['status_final'] = df['delta_final'].apply(status)
```

```python
df.head()
```

```python
df['percent_mudanca'] = abs(((df['30h_d'] + df['06h_d']) / (df['30h'] + df['06h']))-1)
```

```python
df.head()
```

```python
def delta_ajuste(a):
    if a['percent_mudanca'] > 0.4:
        if a['status_final'] == 'Acrescimo':
            return int((a['30h']+a['06h'])*0.4)
        elif a['status_final'] == 'Reducao':
            return -(int((a['30h']+a['06h'])*0.4))
        else:
            return 0
    else:
        return a['delta_final']
```

```python
df['delta_ajustado'] = df.apply(delta_ajuste, axis=1)
```

```python
df.head()
```

```python
def delta_30h_ajustado(a):
    if a['delta_30'] == 0:
        return 0
    elif a['percent_mudanca'] < 0.4:
        return a['delta_30']
    elif a['delta_06'] == 0:
        return a['delta_ajustado']
    elif a['delta_30'] > 0 and a['delta_06'] > 0:
        if a['delta_30'] < a['delta_ajustado']:
            return a['delta_30']
        else:
            return a['delta_ajustado']
    elif a['delta_30'] < 0 and a['delta_06'] < 0:
        if a['delta_ajustado'] < a['delta_06']:
            return (a['delta_ajustado'] - a['delta_06'])
        else:
            return 0 
    elif a['delta_30'] > 0 and a['delta_06'] < 0 and a['delta_ajustado'] > 0:
        return a['delta_30'] - ((a['delta_30']+a['delta_06'])-a['delta_ajustado'])
    elif a['delta_30'] > 0 and a['delta_06'] < 0 and a['delta_ajustado'] < 0:
        return  a['delta_30']
    elif a['delta_30'] < 0 and a['delta_06'] > 0 and a['delta_ajustado'] > 0:
        return  a['delta_30']
    elif a['delta_30'] < 0 and a['delta_06'] > 0 and a['delta_ajustado'] < 0:
        return  a['delta_ajustado'] - a['delta_06']
```

```python
df['delta_30h_ajustado'] = df.apply(delta_30h_ajustado, axis=1)
```

```python
def delta_06h_ajustado(a):
    if a['delta_06'] == 0:
        return 0
    elif a['percent_mudanca'] < 0.4:
        return a['delta_06']
    elif a['delta_30'] == 0:
        return a['delta_ajustado']
    elif a['delta_30'] > 0 and a['delta_06'] > 0:
        if a['delta_30'] >= a['delta_ajustado']:
            return 0
        else:
            return a['delta_ajustado'] - a['delta_30']
    elif a['delta_30'] < 0 and a['delta_06'] < 0:
        if a['delta_ajustado'] < a['delta_06']:
            return a['delta_06']
        else:
            return a['delta_ajustado'] 
    elif a['delta_30'] > 0 and a['delta_06'] < 0 and a['delta_ajustado'] > 0:
        return a['delta_06']
    elif a['delta_30'] > 0 and a['delta_06'] < 0 and a['delta_ajustado'] < 0:
        return  a['delta_06'] - ((a['delta_06']+ a['delta_30']) - a['delta_ajustado'])
    elif a['delta_30'] < 0 and a['delta_06'] > 0 and a['delta_ajustado'] > 0:
        return  a['delta_06'] - ((a['delta_06']+ a['delta_30']) - a['delta_ajustado'])
    elif a['delta_30'] < 0 and a['delta_06'] > 0 and a['delta_ajustado'] < 0:
        return  a['delta_06']
```

```python
df['delta_06h_ajustado'] = df.apply(delta_06h_ajustado, axis=1)
```

```python
df.head()
```

```python
def hall_final1(a):
    if a['minimo'] == "Minimo" and a['status_final'] == "Reducao":
        return a['30h'], a['06h']
    else:
        return a['30h'] + a['delta_30h_ajustado'], a['06h'] + a['delta_06h_ajustado']  
```

```python
df['30h_final1'], df['06h_final1'] = zip(*df.apply(hall_final1, axis=1))
```

```python
def ajuste_fila_23(a):
    if a['30h'] > 0:
        if a['status_30'] == "Acrescimo":
            if a['fila_30'] < 0.25:
                return a['30h'], a['06h_final1']
            elif a['fila_30'] < 0.4:
                return a['30h'], a['06h_final1']
            elif a['fila_30'] < 0.6:
                return a['30h'], a['06h_final1']
            elif a['fila_30'] > 0.6 and (a['dim_22_23'] < 1 and a['saques_21_23'] < -7):
                return a['30h'] + 1, a['06h_final1']
            elif a['fila_30'] > 0.6 and (a['dim_22_23'] > -1 and a['saques_21_23'] > -7):
                return a['30h_final1'], a['06h_final1']
            else:
                return a['30h'] + 1, a['06h_final1']
        if a['status_30'] == "Quadro Ideal":
            if a['fila_30'] < 0.25:
                return a['30h'] - 1, a['06h_final1']
            elif a['fila_30'] < 0.4:
                return a['30h'], a['06h_final1']
            elif a['fila_30'] < 0.6:
                return a['30h'], a['06h_final1']
            elif a['fila_30'] > 0.6 and (a['dim_22_23'] < 1 and a['saques_21_23'] < -7):
                return a['30h'] + 1, a['06h_final1']
            elif a['fila_30'] > 0.6 and (a['dim_22_23'] > -1 and a['saques_21_23'] > -7):
                return a['30h'] + 1, a['06h_final1']
            else:
                return a['30h'] + 1, a['06h_final1']
        if a['status_30'] == "Reducao":
            if a['fila_30'] < 0.25:
                return a['30h_final1'], a['06h_final1']
            elif a['fila_30'] < 0.4:
                return a['30h_final1'], a['06h_final1']
            elif a['fila_30'] < 0.6:
                return a['30h'], a['06h_final1']
            elif a['fila_30'] > 0.6 and (a['dim_22_23'] < 1 and a['saques_21_23'] < -7):
                return a['30h'], a['06h_final1']
            elif a['fila_30'] > 0.6 and (a['dim_22_23'] > -1 and a['saques_21_23'] > -7):
                return a['30h']+1, a['06h_final1']
            else:
                return a['30h']+1, a['06h_final1']
    elif a['06h'] > 0 and a['30h'] == 0:
        if a['status_06'] == "Acrescimo":
            if a['fila_06'] < 0.25:
                return a['30h'], a['06h']
            elif a['fila_06'] < 0.4:
                return a['30h'], a['06h']
            elif a['fila_06'] < 0.6:
                return a['30h'], a['06h']
            elif a['fila_06'] > 0.6 and (a['dim_22_23'] < 1 and a['saques_21_23'] < -7):
                return a['30h'], a['06h']+1
            elif a['fila_06'] > 0.6 and (a['dim_22_23'] > -1 and a['saques_21_23'] > -7):
                return a['30h'], a['06h_final1']
            else:
                return a['30h'], a['06h']+1
        if a['status_06'] == "Quadro Ideal":
            if a['fila_06'] < 0.25:
                return a['30h'], a['06h']-1
            elif a['fila_06'] < 0.4:
                return a['30h'], a['06h']
            elif a['fila_06'] < 0.6:
                return a['30h'], a['06h']
            elif a['fila_06'] > 0.6 and (a['dim_22_23'] < 1 and a['saques_21_23'] < -7):
                return a['30h'], a['06h']+1
            elif a['fila_06'] > 0.6 and (a['dim_22_23'] > -1 and a['saques_21_23'] > -7):
                return a['30h'], a['06h']+1
            else:
                return a['30h'], a['06h']+1
        if a['status_06'] == "Reducao":
            if a['fila_06'] < 0.25:
                return a['30h'], a['06h_final1']
            elif a['fila_06'] < 0.4:
                return a['30h'], a['06h_final1']
            elif a['fila_06'] < 0.6:
                return a['30h'], a['06h']
            elif a['fila_06'] > 0.6 and (a['dim_22_23'] < 1 and a['saques_21_23'] < -7):
                return a['30h'], a['06h']
            elif a['fila_06'] > 0.6 and (a['dim_22_23'] > -1 and a['saques_21_23'] > -7):
                return a['30h'], a['06h']+1
            else:
                return a['30h'], a['06h']+1
    else:
        return 0,0
```
